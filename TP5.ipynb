#### **Exercice 04:** 
>**Ã‰crire une fonction somme avec un argument Â« tuple de longueur variable Â» qui calcule la somme des nombres contenus dans le tuple.Tester cette fonction par des appels avec diffÃ©rents tuples dâ€™entiers ou de flottants.**

```python
def somme_tuple(t):
    """Prend en argument un tuple et calcule la somme de ses Ã©lements"""
    som=0
    for i in t:
        som=som+i
    return som
print(somme_tuple((1,4,4,4)))
print(somme_tuple((1.5,9,6,4,6)))


```

```python
def somme(n):
    """Prend en argument un entier ð‘› et retourne la somme des ð‘›-prÃ©miers entiers naturels."""
    s=0
    for i in range(0,n):
        s+=i
    return s
```
